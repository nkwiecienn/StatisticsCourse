Today's lesson is devoted to testing statistical hypotheses. This is a method of inductive reasoning, or an attempt to draw certain conclusions about the world around us without direct evidence of the relationship. Although this method of reasoning may seem far-fetched, it has been successfully used for centuries in all kinds of cases. A good example of such reasoning are court cases. Of course, if you managed to literally catch the criminal by the hand, such reasoning is unnecessary, but in a significant part of cases we do not have such convenience. Then arguments are presented that are supposed to convince us of the guilt of the accused and if the chances that he is not the culprit are small enough (in an ideal world), he is considered guilty of the crime. The procedure for testing a statistical hypothesis is based on such a model and assuming the role of the accuser in it.

We are given a certain sample (the accused) and certain assumptions about the world that we would like to overturn (the so-called presumption of innocence). These assumptions allow us to establish a certain statistic whose distribution will be known to us (the distribution of the statistic, not the population from which the sample was taken). We can calculate the value of this statistic because we have a sample. What's more, we can calculate the probability of this statistic reaching this value based on our assumptions (this fragment is convincing guilt). If the chances of reaching this value are sufficiently small, we consider our initial assumptions to be unfounded (declaring guilty). However, let's use slightly more formal names.

The basic assumption for a given problem is called the null hypothesis. Depending on the problem, it will lead to statistics with a certain known distribution. The null hypothesis is usually a safer assumption that what we want to show may not be true (presumption of innocence). Then we select the so-called alternative hypothesis for the problem, i.e. what we want to show. As a rule, we have three types of them: left-sided, right-sided and two-way. The first one says that the null hypothesis is incorrect and the value being tested is smaller, the right-sided one claims that it is larger, and the two-way one cannot decide and says that it is simply different. As in the case of estimators, we also decide with what probability we would like to be wrong (here we call it the significance level of the test and denote it by $\alpha$). Based on this significance level and the distribution given by our assumptions, we calculate the so-called critical value, i.e. the value beyond which the probability of obtaining a given value is too small and we can relatively safely reject the null hypothesis in favor of the alternative one. Typicaly statistical tests will give you the $p$ value for a given sample. This is the probability of creating the sample under the assumption that the null hypothesis is true, thus if it is sufficiently small it is the reason to reject the null hypothesis.

The basic test that we want to take a look at is a test that checks is a sample is taken from a normal distribution. As this assumption is important for several cases this is a very popular statistical test, and has several variants.
```r
set.seed(42)
v_norm <- rnorm(500, mean = 6, sd = 3)
v_unif <- runif(500, min = 5, max = 100)

shapiro.test(v_norm)
shapiro.test(v_unif)
```

Other typical uses of hypothesis testing is to check for the parameters of the distribution. One of the basic tests to check for the value of the mean (or the median) is the t.test.

Create a random vector from the normal distribution of size 200, with mean = 5 and sd =10. Test if the mean is equal to 6.

```r
set.seed(42)
v <- rnorm(200, mean = 5, sd = 10)
t.test(v, mu = 6)
```

The t.test can also be used to check is two samples have similar means. We can test wether other parameters of a population ar similar for two populations. Like for variance we can use the var.test.

Create two random vectors from the normal distribution of size 200, with mean = 5 and first sd =6 and second sd=3. Test if the means of the samples are equal. Next check the hypothesis that the sample variances are equal.

```r
set.seed(42)
v1 <- rnorm(200, mean = 5, sd = 6)
v2 <- rnorm(200, mean = 5, sd = 3)

t.test(v1, v2)
var.test(v1, v2)
```

The file Va.csv contains answers to some of the questions from a 2024 survey among data scientists on Kaggle. The first row contains the questions answered in the column. Remove all the rows that have no answer in the column that contains information about saleries. Replace also the intervals there by either the mean value of the interval, or a uniformly drawn value from that interval. Use the data to check the hypothesis that the mean salary for data scientists that use only R is higher than of those that use only Python. Do a similar test for those that use both and those that use only Python. Choose a country. Run a similar test limited to that country.

```r
Va <- read.csv2('Data/Va.csv')
Va <- Va[-1, ]  # remove question-description row
names(Va) <- c('country', 'salary', 'python', 'r_lang')

Va <- Va[Va$salary != '' & Va$salary != 'I do not wish to disclose my approximate yearly compensation', ]

parse_salary <- function(s) {
  s <- gsub(',', '', s)
  if (grepl('\\+', s)) return(as.numeric(gsub('\\+', '', s)))
  parts <- as.numeric(strsplit(s, '-')[[1]])
  mean(parts)
}
Va$salary_num <- sapply(Va$salary, parse_salary)

r_only    <- Va[Va$r_lang == 'R'      & Va$python == '', ]
py_only   <- Va[Va$python == 'Python' & Va$r_lang == '', ]
both_lang <- Va[Va$python == 'Python' & Va$r_lang == 'R', ]

cat('R-only vs Python-only:\n')
print(t.test(r_only$salary_num, py_only$salary_num, alternative = 'greater'))

cat('\nBoth vs Python-only:\n')
print(t.test(both_lang$salary_num, py_only$salary_num, alternative = 'greater'))

Va_us      <- Va[Va$country == 'United States of America', ]
r_only_us  <- Va_us[Va_us$r_lang == 'R'      & Va_us$python == '', ]
py_only_us <- Va_us[Va_us$python == 'Python' & Va_us$r_lang == '', ]

cat('\nUSA – R-only vs Python-only:\n')
print(t.test(r_only_us$salary_num, py_only_us$salary_num, alternative = 'greater'))
```

Bootstraping can also by used in hypothesis testing. But it requires a little more work, as we will have to create the function ourselves. You will have to use the boot function to create a test statistic function based on our starting data and the proper mean function for the test. Then we should find the cumulative distribution function based on the data, what is the quantile of our sample. Based of that we can calculate the $p$ value, which is enough to decide if we reject the null hypothesis.

Use bootstrapping to test whether the mean for the uniform statistic from the first excercise is equal to 20, and do a second test for the mean 45.

```r
library(boot)

mean_fn <- function(data, i) mean(data[i])

bootstrap_test <- function(data, mu0, R = 1000) {
  t_obs <- mean(data)
  data_shifted <- data - t_obs + mu0
  boot_res <- boot(data = data_shifted, statistic = mean_fn, R = R)
  p_val <- mean(abs(boot_res$t - mu0) >= abs(t_obs - mu0))
  cat('H0: mu =', mu0, '| observed mean:', round(t_obs, 3), '| p-value:', p_val, '\n')
}

set.seed(42)
bootstrap_test(v_unif, mu0 = 20)
bootstrap_test(v_unif, mu0 = 45)
```
