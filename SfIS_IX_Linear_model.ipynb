Today we will focus on building linear models to analyze a given variable. The methods used differ depending on the type of variable you are using to describe the data (the variable used to describe is called the independent variable). We will start with analyzing a quantitative variable using a qualitative one.


For that we use the method known as Analysis of Variance (ANOVA for short). The idea here is simple. While obviously we cannot find a linear equation between the quantitative data and qualitative data because of discrete nature of the latter (and probable lack of natural order on this variable), we still can check how the variable in question bechaves when grouped by the qualitative one. One we have this grouping the null hypothesis becomes that the mean values between the groups are roughly the same (the alternative is that for some this is not true). To test if this hypothesis holds true we have to create a linear model between the variables. For that we use the function lm(). When analyzing against only one variable (which is called one way analysis) we put the formule in the for main_valiable~independent_variable to this function. On this linear model we can use the the function anova() (or the aov() function, there is some small difference in the way the results are presented but it does more less the same) to check if the null hypothesis holds any ground. If there is a reason to reject this hypothesis there will be a number of * near the p-value (the more * the more significant difference.)

If we found that there is a reason to reject the null hypothesis that does not mean that all of the mean values differ. To find out which pairs of groups have different means we conduct what is called a post-hoc test. In our case we will go with Tukey's HSD test (short for Honestly Significant Differences). For that we put the model into the TukeyHSD() function. It will compare the groups pairwise and by the p-values we can see where the differences are significant.

Perform a TukeyHSD test on the aov data from the previous two exercises.

```r
TukeyHSD(model)
```

Of course we can want to group our data by more than one qualitative variable in one time. We could make a one way ANOVA for each of those variables but that way we do not catch any of the dependencies between the qualitative variables. So we conduct what we call multi way ANOVA (two way if we only use two variables). The main difference is how we provide the formula for the lm() function. It is in the form main_variable~independent_variable1(special symbol)independent_variable2, where the (special symbol) is chosen depending on what relation we want to check between the two variables. So we use '+' if we assume no interaction between those variables ':' if we want to focus only on the interaction and '*' if we want to do both.

The file Pri.csv contains data about the dependence between the price of apartments and several parameters in 3 different districts. Analyse the variance one way against the district and the type of building, and then conduct a twoway analysis against both of those variables at the same time, without assuming interaction.

```r
C <- read.csv2("Data/Pri.csv")
C$district <- as.factor(C$district)
C$type <- as.factor(C$type)
C$price <- as.numeric(C$price)
C
```

```r
model1 <- aov(price ~ district, data = C)
```

```r
model2 <- aov(price ~ type, data = C)
```

```r
model3 <- aov(price ~ district + type, data = C)
```

```r
shapiro.test(residuals(model1))
```

```r
shapiro.test(residuals(model2))
```

```r
shapiro.test(residuals(model3))
```

```r
leveneTest(price ~ district, data = C)
```

```r
leveneTest(price ~ type, data = C)
```

```r
leveneTest(price ~ district * type, data = C)
```

```r
summary(model1)
```

```r
summary(model2)
```

```r
summary(model3)
```

Things get a little simpler if the variable we are using to describe the date we are interested in is of a quantitative kind. In such a case we can assume that the data can be described as a linear function of the variable we are using (whether this is reasonable highly depends on the covariance between the variables). Such an approach we call linear regression and the lm() function will straight out calculate the coefficients of the most probable linear function to describe this relation. They will be given in the model under the value 'coefficients'.





The file Per.csv contains the heights of baby brothers and sisters in different families. Construct a linear regression model of the height of boys against girls. Plot the model.

```r
library(e1071)
Per <- read.csv2('Data/Per.csv')
Per_complete <- na.omit(Per)

model_per <- lm(brother ~ sister, data = Per_complete)
summary(model_per)

plot(Per_complete$sister, Per_complete$brother,
     xlab = 'sister', ylab = 'brother',
     main = 'brother ~ sister')
abline(model_per, col = 'red')
```

If we have the formula for the regression line we can use it to predict the value of the data we are interested in by simply supplying the values of the independent variable to the formula for the regression line.

The file Books.csv contains the data of the relation between numbers of books read and the final spelling scores for third graders. Use it to predict the score for someone who read: 3 books, 7 books, and 49 books.

```r
Books <- read.csv2('Data/Books.csv')
model_books <- lm(spelling ~ book, data = Books)
summary(model_books)

predict(model_books, newdata = data.frame(book = c(3, 7, 49)))
```

As in the case of ANOVA we can test our data against multiple independent quantitative variables. The way we provide the formula is similar to what we did in two way ANOVA.

Use the data from Pri.csv for this exercise to create a regression model for price against the area of the apartment, the number of rooms separately, and against both the area and number of rooms.

```r
model_area  <- lm(price ~ area,         data = C)
model_rooms <- lm(price ~ rooms,        data = C)
model_both  <- lm(price ~ area + rooms, data = C)

summary(model_area)
summary(model_rooms)
summary(model_both)
```

Finally it is somewhat obvious that not always we know which variables are significant to describing the data we are interested in. We also do not want to check each combination of variables by ourselves. Luckily we have the function step() that can check that for us. This function has an important parameter called 'direction'. If it is forward than the function will try to add new parameters if they are useful to the model so in this case we want to start with as small model as possible. If backward it will try to subtract independent variables, checking if the loss of information is significant, so we want to start with a model that has all of the variables that we want to check.  

Read the documentation to the step() function and use it to decide if the model with both area and number of rooms are necessary for a good model of price.

```r
step(model_both, direction = 'backward')
```
