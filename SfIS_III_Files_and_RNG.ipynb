Generating pseudorandom numbers is often important were statistics are involved. In R several algoritms to generate those numbers are available and you have always two generators ready when you call a function that uses randomness (one generator for the uniform distribution, one for the normal distribution). You can use `RNGkind(uniform algo, normal algo)` to change the algorithms, and you will get the names of the previously used as output. As you may know, when generating pseudorandom numbers you use a seed with the values given to the algorithm. It is kept under the variable `.Random.seed`. You can set the starting point of the seed using the function `set.seed(value)`.

There are several functions for generating random values and other useful information from known distributions. The have a common naming convention: $[prefix][name of distribution]$ where prefix is one of the four: r, p, d, q, and the name is usually a four letter shortcut. r- is for generating random numbers from the distribution, p- gives the values from the cumulative distribution function, d gives  the value of density in that point for contunous distributions and the probability of given values for discreate distributions, and q is used to find quartiles. norm is for normal distribution, unif for uniform and chisq for chi-squared.
