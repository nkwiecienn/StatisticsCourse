The focus of this lab is around the package `lattice`. It is a package that focuses on drawing certain kinds of graphs. While the basic graphics package allows you to draw any graph possible with sufficent knowledge, lattice has several advantages over graphics. For example all functions have a smilar structure, which makes using them easier, a lot of default settings are selected to make the graphs more readable than in the standard graphics package, and also some tasks, like creating a legend for the graph, are automated.

The basic input to any function in lattice is in a form of a formula `L~ R | D` where `L` has all variables you wish to present on the y axis, `R` all variables you wish to present on the x axis, and `D` has all of the variables you want to use as dependencies (is none you should drop the | symbol). you can use multiple variables in those parts, you have to separate the values with the + symbol. Also sometimes we do not need one of those variables at all, for example in a histogram we just need a variable for the x axis, so we call out the function in the form `histogram(~x)` where `x` is the variable we want a histogram of.

To start our work, file ap.csv which contains data from the variable apartments from the PBImisc library, and contains information about appartment offers in Warsaw in years 2007-2009. As the first task draw a histogram (using the function from the package lattice) of the percentage of offers depending on the district in which the offer is placed. Next draw the histogram for those districts in which the percentage of offers exceeded 3 percent (you may need to modify your data frame for this)

```r
ap = read.csv2('Data/ap.csv')
ap
```

```r
library(lattice)

district_counts <- as.data.frame(table(ap$district))
names(district_counts) <- c("district", "n")

district_counts$percent <- 100 * district_counts$n / sum(district_counts$n)

histogram(~district_counts$percent)
```

```r
district_counts$percent
```

```r
library(lattice)

district_above3 <- subset(district_counts, percent > 3)

histogram(~district_above3$percent)
```

There are a total of 15 functions in the lattice package. The next one we will be looking in is the `bwplot()` which creates a box whiskers plot. Now we will need both a dependent variable(`y`) (preferably numeric), and a independent variable(`x`) (for a box whiskers plot preferably a factor) in our formula. For the next task try to create this kind of graph for the prices of offers for apartaments depending on the district they are from (you may use the reduced data for clearer view).

```r
library(lattice)

bwplot(
  district ~ offer.price,
  data = ap
)
```

```r
library(lattice)

district_counts <- table(ap$district)

selected_districts <- names(district_counts[district_counts >= 30])

ap_reduced <- subset(ap, district %in% selected_districts)

bwplot(
  district ~ offer.price,
  data = ap_reduced,
)
```

If both variables we are interested in are quantitative in their nature it is best to draw a scatterplot to represent the data. For that we have the `xyplot()` function. First try to draw a scatterplot of the relation between the offer price and the actual transaction price for our data (we can use all data points in this case). Next we may want to check what was this relation but we want to see the relation depending on the condition of the apartment on separate graphs. This is the point where we can use a formula with the dependencies variable, the rest will be done automatically by the functions of lattice.


```r
library(lattice)

xyplot(
  offer.price ~ transaction.price | condition,
  data = ap
)
```

As we see if we use a dependencies variable we will get separate graphs for each situation. What could we do if we wanted them distingushable on one graph instead? In that case instead of altering the formula we can use the parameter "`group`" in our function to specify on which variable should be used to distingush the scatter points on our graph. As the points will be now distingushable depending on what type they are, the graph will need a legend. Read about the parameter "auto.key" to see how to add one quickly. Repeat the last task but make so that all points are on one graph distingushable on thier condition with an added legend.

```r
library(lattice)

xyplot(
  offer.price ~ transaction.price,
  group = condition,
  auto.key = TRUE,
  data = ap,
)
```

Sometimes we want to take several variables an draw graphs of their relations for each pair. Instead of doing this for each pair separatly, we can use the function `splom()`. Use this function to draw in one graph all the pairwise relations between the number of rooms, the floor, the price of a square m, and the price of transaction.

```r
library(lattice)

splom(
  ~ n.rooms * floor * m2.price * transaction.price,
  data = ap
)
```

We can add more information to our plots by using the "type" parameter in our functions. Read about it and add to all of our scatterplots the regression line for that plot and a grid to make the graph more easy to read.

```r
library(lattice)

xyplot(
  offer.price ~ transaction.price,
  group = condition,
  auto.key = TRUE,
  type = c("p","r"),
  grid = TRUE,
  data = ap,
)
```
