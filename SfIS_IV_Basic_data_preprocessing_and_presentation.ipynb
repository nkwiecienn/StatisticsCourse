This lab focuses around data frames, data preprocessing and basic descriptive statistics.


In most cases, when working on statistical evaluation of data, in R we work with a data structure called a dat frame. It is effectively a table with (usually) named columns, in which values are of the same data type, and (sometimes) named rows, which represent one element of the sample. Thus in one row we have all the values given from one specific instance of the population. As a table you can call specific cells of the table by directly adressing them in the format `dataFrame[vector_of_rows,vector_of_columns]`. If you leave one of the vectors empty it will call all of the values, so `A[2:3,]` will give all of the values of row 2 and 3 in data frame `A`. If the columns are named we can address the directly by using the names of the columns in the form `dataFrame$name_of_column` which will give a vector with the values of the column.


When working with real data it is often the cases that some of the values are missing for some observations. We can check for cells with missing values using `is.na()` function, which returns TRUE if there is a value, and FALSE otherwise. If we want to ignore the missing values in a vector we have the `na.omit()` function, and if we want only the rows with all values in a data frame we use `complete.cases()`.


Sometimes, if we would use only complete cases too much of the data would be lost. It is sometimes reasonable to impute values to replace the missing ones. There are simple functions to do just that. In library `Hmisc` there is the function `impute(data,fun)`, where data is a vector or data frame with missing values, and fun is what we use for replacements. It can be a specific value, `mean` if you want to replace with the mean value of the column, or `"random"` if you want to replace with random numbers take directly from te sample. Another function that can be used is `impute()`  from the library `e1071`, which can also replace with the median of the sample. If you have two functions with the same name from different librries in R both are avalable to you and the default is the one from the last library to be loaded. If you want to call a specific function you can write `name_of_library::name_of_function`.


If we want to make changes to a column, or columns in a data frame we can use `transform(data_frame, name_of_column = function(name_of column))`. As when loading  data freme, R automatically assumes the data type of the column, this can be used to change a numeric type to a factor type using `as.factor()` (factor is a qualitative data type). In general `as.data_type()` will try to convert a vector to a given data type. We can also use `scale()` to normalize the data in a column.



Exercise 1. The file Chem.csv contains data on starting salaries for chemists, with missing data. Complete the data
once using random values from the data, and a second time using the mean values. Save the data to variables A and B
respectively.

```r
install.packages('Hmisc')
library('Hmisc')

Chem = read.csv2('Data/Chem.csv')
A <- data.frame(salary = as.numeric(impute(Chem$salary, fun=mean)))
B <- data.frame(salary = as.numeric(impute(Chem$salary, fun='random')))
```

Descriptive statistics is focused on presenting raw data in a form that is more interpretable for humans. Raw numbers do not present a clear picture, but with a little knowledge parameters like `mean()`, `median()`, will give us some information about wher the data is centered, while `range()`, Inter Quartile Range `IQR()`, standard deviation `sd()`, or variance `var()` gives us information about the dispersion of data. We can call for this information separately, or use the function `summary()` for a full rundown.
